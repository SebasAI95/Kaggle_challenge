Titanic Challenge Description

    Main idea is use Titanic passenger data to train, fit model and predict whether each passenger survived or not.
    Titanic passenger data :
    | PassengerId | #Survived | #Pclass | Name | Sex | # Age | SibSp | Parch | Ticket | Fare |

    In previous columns, Survived (yes : 1, not : 0) is predicted LABEL

Here, column descriptions are into :
https://databricks-prod-cloudfront.cloud.databricks.com/public/4027ec902e239c93eaaa8714f173bcfc/19095846306138/45566022600459/8071950455163429/latest.html#:~:text=aboard%20the%20Titanic-,parch,fare


```python
# Required libraries
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
```

```python
# First, load train dataset
train_data = pd.read_csv("resources/train.csv")
train_data.head()
```

```python
# First, load test dataset
test_data = pd.read_csv("resources/test.csv")
test_data.head()
```

Exploratory Data Analysis

Exist initial hypothesis from Kaggle that all female passengers survived, so is possible
calculate corresponding %.

```python
# female - male %
male_percent = train_data[train_data["Sex"] == "male"]["Survived"].value_counts(normalize=True)
female_percent = train_data[train_data["Sex"] == "female"]["Survived"].value_counts(normalize=True)

print("female passengers survived are :", female_percent.loc[1])
print("male passengers survived are :", male_percent.loc[1])

# Initial hypothesis is no true, so 25 % of female passengers have died, although is good estimation.
# From male passengers is much worse, so only 19% have survived.
```

So, estimate if passenger live/dead only from sex is not the best solution, so
only takes into account one column, so is necessary create estimation from multiple columns, so
in this case is complex analyze ny ourselves, so is useful create machine learning model
that can learn from training data common patterns.

Here, i have used to start, random forest model, that's several decision trees that predict individually
and final result is done from democratic decision.




```python
# Random forest model
features = ["Pclass", "Sex", "SibSp", "Parch"]
X_train = pd.get_dummies(train_data[features])
y_train = train_data["Survived"]
X_test = pd.get_dummies(test_data[features])

model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
# Now, create .csv file with | PassengerId" | Survived | for test dataset including
# y_pred to upload into platform and evaluate accuracy.

output_df = pd.DataFrame({"PassengerId" : test_data.PassengerId, "Survived" : y_pred})
output_df.to_csv("submission.csv", index=False)
```
